```python
import scipy
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
x = np.linspace(-1, 1, 250)
y = x**2 * np.sin(1/x**2) + x
plt.xlabel('Ось x')
plt.ylabel('Ось y')
plt.title('График')
plt.grid()
plt.plot(x, y, label = 'x^2*sin(1/x^2)+x')
plt.legend()
```

```python
x = np.linspace(5, 25, 75)
a = 1 / (1 + x ** 2)
b = 1 / x ** 2
y = abs(1 - a / b)
plt.title('График 2')
plt.xlabel('Ось x')
plt.ylabel('Ось y')
plt.semilogy(x, y)
```

```python
x = np.linspace(-5, 5, 50)
y = scipy.special.gamma(x) / 25 # на максимум делить не хочет, поэтому нормировка условная
y2 = scipy.special.psi(x) / 50
y3 = scipy.special.erf(x)
plt.title('График 3')
plt.xlabel('Ось x')
plt.ylabel('Ось y')
plt.plot(x, y, label = 'нормированная Гамма-Функция')
plt.plot(x, y2, label = 'нормированная Пси-Функция', color = 'g', linestyle = '--')
plt.plot(x, y3, label = 'функция ошибки', color = 'r', linestyle = '', marker = '*')
plt.legend()
```

## Exercise 2  
1. Построить график функции $sin(x)/x$ на отрезке $[-6\pi;6\pi]$ и добавить тики по оси $x$ с шагом $\pi/2$ от $-6\pi$ до $6\pi$, причём обозначения тиков должны содержать греческую букву $\pi$. Добавить сетку на график.

```python
x = np.linspace(-6 * np.pi, 6 * np.pi, 250)
y = np.sin(x) / x
plt.figure(figsize=(12, 6))
plt.xlabel('Ось x')
plt.ylabel('Ось y')
xt = np.arange(-6 * np.pi, 6 * np.pi + np.pi/2, np.pi/2)
yt = []
for i in range(-12, 13):
    if i % 2 == 0:
        if i == 0:
            yt.append('0')
        elif i == -2:
            yt.append(r'$-\pi$')
        elif i == 2:
            yt.append(r'$\pi$')
        else:
            yt.append(r'$%s\pi$'%(i // 2))
    else:
        if i != 1 and i != -1:
            yt.append(r'$\dfrac{%s\pi}{2}$'%(i))
        elif i == -1:
            yt.append(r'$-\dfrac{\pi}{2}$')
        else:
            yt.append(r'$\dfrac{\pi}{2}$')
plt.xticks(xt, yt)
plt.grid()
plt.plot(x, y)
```

## Exercise 3  
1. Создайте цветовой график $sin(x)sin(y)$ на интервале $[−π,π]×[−π,π]$.
2. Создайте цветовой график $e^{-x^2-y^2}$ на интервале $[−3,3]×[−3,3]$.

```python
import matplotlib as mpl

x = np.linspace(-np.pi, np.pi, 250)
y = np.linspace(-np.pi, np.pi, 250)
X,Y = np.meshgrid(x,y)

Z = np.sin(X) * np.sin(Y)

fig, ax = plt.subplots(figsize=(6,5))

norm = mpl.colors.Normalize(-abs(Z).max(), abs(Z).max())

p = ax.pcolor(X, Y, Z, norm = norm, cmap=mpl.cm.Paired)
ax.axis('tight')
ax.xaxis.set_major_locator(mpl.ticker.MaxNLocator(4))
ax.yaxis.set_major_locator(mpl.ticker.MaxNLocator(4))

cb = fig.colorbar(p, ax=ax)
cb.set_ticks([-1, -.5, 0, .5, 1])
```

```python
import matplotlib as mpl

x = np.linspace(-3, 3, 250)
y = np.linspace(-3, 3, 250)
X,Y = np.meshgrid(x,y)

Z = np.exp(- X**2 - Y**2)

fig, ax = plt.subplots(figsize=(6,5))

norm = mpl.colors.Normalize(-abs(Z).max(), abs(Z).max())

p = ax.pcolor(X, Y, Z, norm = norm, cmap=mpl.cm.Paired)
ax.axis('tight')
ax.xaxis.set_major_locator(mpl.ticker.MaxNLocator(4))
ax.yaxis.set_major_locator(mpl.ticker.MaxNLocator(4))

cb = fig.colorbar(p, ax=ax)
cb.set_ticks([-1, -.5, 0, .5, 1])
```
