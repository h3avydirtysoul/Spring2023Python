```python
import numpy as np
import numpy.linalg
import numpy.random
import matplotlib.pyplot as plt
```

# Номер 1

```python
def f(x):
    return np.sin(x/5) * np.exp(x/10) + 5 * np.exp(-1/2 * x)


def approximation(n):
    a = np.linspace(1, 15, n)
    A = np.array([[x ** i for i in range(n)] for x in a])
    B = f(a)
    W = np.linalg.solve(A, B)
    answer = np.array([sum([W[i] * x**i for i in range(n)]) for x in a])
    return answer


x = np.linspace(1, 15, 250)
n = 5
x2 = np.linspace(1, 15, n)
plt.grid()
plt.xlabel('Оx')
plt.ylabel('Оy')
plt.plot(x, f(x), label = 'Исходная функция')
plt.plot(x2, f(x2), marker = '.', linestyle = '', label = f'Точки для аппроксимации при n = {n}')
plt.legend()
plt.show()

plt.grid()
plt.xlabel('Ось x')
plt.ylabel('Ось y')
plt.plot(x2, approximation(n), label = f'Аппроксимация при n = {n}')
plt.plot(x2, f(x2), marker = '.', linestyle = '', label = f'Точки для аппроксимации при n = {n}')
plt.legend()
plt.show()
```

# Номер 2

```python
def noize_approximation(n):
    a = np.linspace(1, 15, n)
    A = np.array([[x ** i for i in range(n)] for x in a])
    B = f(a)
    for i in range(len(B)):
        B[i] += np.random.normal(0, 0.01)
    W = np.linalg.solve(A, B)
    answer = np.array([sum([W[i] * x**i for i in range(n)]) for x in a])
    return answer, B


x = np.linspace(1, 15, 250)
n = 5
x2 = np.linspace(1, 15, n)
y = noize_approximation(n)
plt.grid()
plt.xlabel('Ось x')
plt.ylabel('Ось y')
plt.plot(x, f(x), label = 'Исходная функция')
plt.plot(x2, y[1], marker = '.', linestyle = '', label = f'Точки для аппроксимации при n = {n}')
plt.legend()
plt.show()

plt.grid()
plt.xlabel('Ось x')
plt.ylabel('Ось y')
plt.plot(x2, y[0], label = f'Аппроксимация при n = {n}')
plt.plot(x2, y[1], marker = '.', linestyle = '', label = f'Точки для аппроксимации при n = {n}')
plt.legend()
plt.show()
```

При увеличении n, наш график становится другим, поэтому делаем вывод, что полиномиальная регрессия с зашумлённым данными лучше работает при малых n.

```python
def f(x):
    return np.sin(x/5) * np.exp(x/10) + 5 * np.exp(-1/2 * x)


def least_squares(n):
    N = 100
    a = np.linspace(1, 15, N)
    A = np.array([[x ** i for i in range(n)] for x in a])
    B = f(a)
    for i in range(len(B)):
        B[i] += np.random.normal(0, 0.1)
    W, unnes1, unnes2, unnes3 = np.linalg.lstsq(A, B, rcond=None)
    answer = np.array([sum([W[i] * x**i for i in range(n)]) for x in a])
    return answer, B


x = np.linspace(1, 15, 250)
N = 100
x2 = np.linspace(1, 15, N)
n = 4 
y = least_squares(n)
plt.grid()
plt.xlabel('Ось x')
plt.ylabel('Ось y')
plt.plot(x, f(x), label = 'Исходная функция')
plt.plot(x2, y[1], marker = '.', linestyle = '', label = f'Точки для аппроксимации при n = {n}')
plt.legend()
plt.show()

plt.grid()
plt.xlabel('Ось x')
plt.ylabel('Ось y')
plt.plot(x2, y[0], label = f'Аппроксимация при n = {n}')
plt.plot(x2, y[1], marker = '.', linestyle = '', label = f'Точки для аппроксимации при n = {n}')
plt.legend()
plt.show()
```

Полиномиальная регрессия хорошо аппроксимирует функции при больших n и при отсутствии сильного шума, метод наименьших квадратов же аппроксимирует довольно точно и при маленьких n и сильном шуме
